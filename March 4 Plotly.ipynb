Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.

```python
import seaborn as sns
import plotly.express as px

titanic = sns.load_dataset('titanic')

figure = px.scatter(titanic, x = 'age', y = 'fare', color = 'class')

figure.show()
```

Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
import plotly.express as px
tips = px.data.tips()

figure = px.box(tips, x = "day", y = "total_bill", color = "smoker", notched = True)

figure.show()
```

Q3: Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.

```python
import seaborn as sns
import plotly.express as px

tips = sns.load_dataset("tips")

figure = px.histogram(tips, x = "sex", y = "total_bill", pattern_shape = "smoker", color = "day", marginal = "box", nbins = 30, opacity = 0.7, 
                      labels = {"sex":"Sex", "total_bill": "Total Bill", "day":"Day", "smoker":"Smoker"})

figure.show()
```

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.
Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
dimensions parameter.

```python
import plotly.express as px
from sklearn.datasets import load_iris

iris = load_iris()
df = px.data.iris()

fig = px.scatter_matrix(
    df,
    dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
    color="species",
    title="Scatter Matrix Plot for Iris Dataset",
    labels={col: col.replace('_', ' ').title() for col in df.columns}
)

fig.show()
```

Q5. What is Distplot? Using Plotly express, plot a distplot.

Distplot is a plot used to visualize the distribution of a dataset. It combines a histogram and a kernel density plot to represent the distribution of data.

```python
import plotly.express as px
import seaborn as sns
tips = sns.load_dataset("tips")

fig = px.histogram(tips, x="total_bill", nbins=20, opacity=0.7, marginal="rug", 
                   histnorm="probability density", color_discrete_sequence=['#EF553B'])
fig.show()
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
colors = ['#FF0000', '#00FF00', '#0000FF', '#FF00FF', '#00FFFF']

source = ColumnDataSource(data=dict(x=x, y=y, colors=colors))

output_file("rect_glyph.html")

p = figure(title="Rect Glyph Example", x_axis_label='X', y_axis_label='Y')

# Add a rect glyph
p.rect(x='x', y='y', width=0.9, height=0.9, fill_color='colors', line_color=None, source=source)
show(p)
```
