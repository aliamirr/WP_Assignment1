Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a popular data visualization library in Python that is widely used for creating various types of graphs, charts, and plots. It provides a range of tools for creating high-quality 2D and 3D visualizations with minimal code.

Here are five examples of plots that can be created using the pyplot module of Matplotlib:

1) Line Plot: A line plot is used to display the relationship between two variables on a continuous axis. It is created using the plot() function in pyplot.

2) Scatter Plot: A scatter plot is used to display the relationship between two variables by plotting points on a 2D plane. It is created using the scatter() function in pyplot.

3) Bar Plot: A bar plot is used to compare the values of different categories using rectangular bars. It is created using the bar() function in pyplot.

4) Histogram: A histogram is used to show the distribution of a numerical variable. It is created using the hist() function in pyplot.

5) Pie Chart: A pie chart is used to show the proportions of different categories in a dataset. It is created using the pie() function in pyplot.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

A scatter plot is a type of plot used to display the relationship between two continuous variables. It consists of a set of data points, each representing a single observation, plotted on a two-dimensional graph.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x, y)
plt.title('Scatter Plot of x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

The subplot() function is used to create multiple plots in a single figure. It allows us to specify the number of rows and columns of subplots in the figure, as well as the index of the current subplot we want to create. This is useful when we want to compare multiple plots side by side or when we want to create a complex visualization that consists of multiple plots.

```python
import numpy as np
import matplotlib.pyplot as plt

#For line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

#For line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

#For line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

#For line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with 2 rows and 2 columns of subplots
figure, axis = plt.subplots(2, 2, figsize=(5, 5))

# Plot line 1 in the top left subplot
axis[0, 0].plot(x1, y1)
axis[0, 0].set_title('Line 1')

# Plot line 2 in the top right subplot
axis[0, 1].plot(x2, y2)
axis[0, 1].set_title('Line 2')

# Plot line 3 in the bottom left subplot
axis[1, 0].plot(x3, y3)
axis[1, 0].set_title('Line 3')

# Plot line 4 in the bottom right subplot
axis[1, 1].plot(x4, y4)
axis[1, 1].set_title('Line 4')

# Add a main title for the whole figure
fig.suptitle('Four Line Plots Using Subplots')

# Show the plot
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

A bar plot is a type of plot that displays data using rectangular bars. The length or height of each bar is proportional to the value of the data it represents. Bar plots are used to compare and display discrete sets of data, and they are commonly used in finance, economics, and other fields to display data such as stock prices, sales figures, and profit margins.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical bar plot
plt.bar(company, profit)
plt.title("Company Profit")
plt.xlabel("Company")
plt.ylabel("Profit ($)")
plt.show()

# Horizontal bar plot
plt.barh(company, profit)
plt.title("Company Profit")
plt.xlabel("Profit ($)")
plt.ylabel("Company")
plt.show()
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

A box plot, also known as a box-and-whisker plot, is a type of plot that displays the distribution of a dataset. It shows the median, quartiles, and outliers of the data in a compact and easily readable format. Box plots are used to compare the distribution of multiple datasets and identify potential outliers or differences in variability.

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a figure and axis
figure, axis = plt.subplots()

# Create the box plot
box_plot = axis.boxplot([box1, box2])

# Set the title and axis labels
axis.set_title('Box Plot')
axis.set_xlabel('Box')
axis.set_ylabel('Value')

# Show the plot
plt.show()
```
