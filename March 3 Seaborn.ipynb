```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
```

Seaborn is a data visualization library built on top of Matplotlib that provides an easy-to-use interface for creating a variety of statistical graphics. Here are five types of plots that can be created using Seaborn along with their uses:

1) Scatter plots: Scatter plots are used to visualize the relationship between two continuous variables. Seaborn's scatter plot function allows for easy customization of the plot by adding color and size to points based on a third variable.

2) Line plots: Line plots are used to visualize trends over time or between different groups. Seaborn's line plot function allows for easy plotting of multiple lines on the same plot with automatic color selection.

3) Bar plots: Bar plots are used to compare discrete categories. Seaborn's bar plot function can be used to create vertical or horizontal bar plots and can automatically aggregate data within categories.

4) Histograms: Histograms are used to visualize the distribution of a single variable. Seaborn's histogram function provides additional options such as kernel density estimation and automatic bin selection.

5) Heat maps: Heat maps are used to visualize the relationship between two categorical variables. Seaborn's heatmap function can also be used to visualize correlation matrices.

Seaborn also provides other types of plots such as violin plots, box plots, and swarm plots for more specialized applications.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns

fmri = sns.load_dataset("fmri")
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
```

```python
Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.
```

```python
import seaborn as sns
titanic = sns.load_dataset("titanic")

# Create the box plots
sns.boxplot(x="pclass", y="age", data=titanic)
sns.boxplot(x="pclass", y="fare", data=titanic)
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
diamonds = sns.load_dataset('diamonds')

#Histogram with hue
sns.histplot(data=diamonds, x="price", hue="cut")

plt.title("Distribution of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Count")
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns
iris = sns.load_dataset('iris')
sns.pairplot(data=iris, hue="species")
```

```python
Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
flights = sns.load_dataset('flights')

# Pivot the data to create a matrix of monthly passenger counts
flights = flights.pivot(index='month', columns='year', values='passengers')

#Heatmap
sns.heatmap(data=flights, cmap='coolwarm', annot=True, fmt='d')

plt.title("Monthly Passenger Counts (1949-1960)")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
